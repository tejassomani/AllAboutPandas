```python
## Histogram

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1, ax2, ax3, ax4]

#distribution for n = 10, 100, 1000, 10000 sample size
for n in range(0,4):
    ## take squares
    sample_size = 10**(n+1) 
    sample = np.random.normal(loc=0.0, scale=1.1, size=sample_size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title("n = {}".format(sample_size))
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gsp = gspec.GridSpec(3,3)

top_right = plt.subplot(gsp[0, 1:])
bottom_left = plt.subplot(gsp[1:4, 0])
bottom_right = plt.subplot(gsp[1:, 1:])

X = np.random.normal(loc=0.0, scale=1.0, size=1000)
Y = np.random.random(size=1000)

bottom_right.scatter(X, Y)
top_right.hist(X, bins=100)
bottom_left.hist(Y, bins=100, orientation='horizontal')


```

```python
#### Box Plots
import pandas as pd

normal = np.random.normal(loc=0.0, scale=1.0, size=1000)
random = np.random.random(size=1000)
gamma = np.random.gamma(2, size=1000)
```

```python
df = pd.DataFrame({'normal': normal, 'random':random, 'gamma': gamma})
```

```python
plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')
```

```python
plt.figure()
plt.hist(df['normal'], bins=100)
```

```python
## BiVariate plots

mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])

sns.jointplot(x="x", y="y", data=df);
```
