```python
##Now cleanup wage data to be numeric 
df.Wage.head()
```

```python
def cleanWage(col):
    c = str(col)
    c = c.replace("€", "")
    if 'K' in c:
        c = c.replace("K", "")
        return float(c) * 1000
    if 'M' in c:
        c = c.replace("M", "")
        return float(c) * 1000000
    return 0
```

```python
df['Wage_in_Euro'] = df['Wage'].apply(cleanWage)
df['Total_Contract_Value'] = df['Value'].apply(cleanWage)
```

```python
#df.info()
df.select_dtypes(include=[int, float])
```

```python
def add_values(col):
    print(type(col))
    c = str(col)
    if '+' in c:
        s = c.split('+')
        return (int(s[0]) + int(s[1]))
    elif '-' in c:
        s = c.split('-')
        return (int(s[0]) - int(s[1]))
    else:
        return int(c)
```

```python
###change column data types to int

to_change = ['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control', 'Composure', 'Crossing', 'Curve', 'Dribbling', 'Finishing', 
'Free kick accuracy', 'GK diving', 'GK handling', 'GK kicking', 'GK positioning', 'GK reflexes', 'Heading accuracy', 'Interceptions', 
'Jumping', 'Long passing', 'Long shots', 'Marking', 'Penalties', 'Positioning', 'Reactions', 'Short passing', 'Shot power', 
'Sliding tackle', 'Sprint speed', 'Stamina', 'Standing tackle', 'Strength', 'Vision', 'Volleys']

for c in to_change:
    df[c] = df[c].apply(add_values)
```
