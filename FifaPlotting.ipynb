```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from matplotlib import rcParams

pd.set_option('display.max_columns', 500)
rcParams.keys()
rcParams.update({'figure.autolayout': True})
```

```python
df = pd.read_csv('Fifa/CleanedPlayers.csv')
```

```python
def plot_histogram(col, xlabel, ylabel, title, size=10, bins=30):
    fig = plt.figure()

    plt.hist(df[col], bins=bins, edgecolor='b')
    plt.xlabel(xlabel, fontsize=size)
    plt.ylabel(ylabel, fontsize=size)
    plt.title(title, fontsize=size)
```

```python
plot_histogram('Age', 'Age', 'No of Players', 'Player Age distribution')
```

```python
print(df['Age'].describe())
fig = plt.figure()
plt.boxplot(df['Age'], whis='range')
```

```python
def get_first_position(col):
    c = str(col)
    pos = c.split(" ")
    return pos[0]

df['player_position'] = df['Preferred Positions'].apply(get_first_position)
```

```python
df['player_position'].unique()
```

```python
def scatter_plot(x_axis, y_axis, x_label, y_label, title, category=None):
    fig, ax = plt.subplots(figsize=(10, 5))
    
    #alpha to add more transparency to the overlapping points
    #if category is not None:
    #colors = ['#2300A8', '#00A658', '#CD5C5C', '#E9967A', '#F2D7D5']
    ax.scatter(df[x_axis], df[y_axis], alpha=0.8, s=40)
    #else:
        #ax.scatter(df[x_axis], df[y_axis], alpha=0.8, s=40)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    
    ax.set_title(title)
    
    ## Remove top and right borders
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    plt.show()
```

```python
scatter_plot('Age', 'Wage_in_Euro', 'age', 'wage (Euro)', 'Age vs wage distribution', 'player_position')
```
